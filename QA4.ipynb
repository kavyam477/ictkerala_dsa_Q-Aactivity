```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data=pd.read_excel("Telco-Customer-Churn.xlsx")
```

```python
data
```

# Carry out Visualization for the following questions:

# 1.Compare churn count with respect to gender.

```python
plt.figure(facecolor="pink",figsize=(10,6))
sns.countplot(data=data,x='gender',hue='Churn',hue_order=["Yes","No"])
plt.title("Churn count with respect to gender",fontsize=20,color="red")
plt.show()
```

# 2. Find out how many female senior citizens there in the dataset 

```python
slice_data=data[(data["SeniorCitizen"]==1) & (data["gender"]=="Female")]
data1=sns.countplot(slice_data["gender"],color="green")
plt.title("Female Senior Citizens",fontsize=18)
```

# 3. Compare 'tenure' with 'Total Charges' 

```python
plt.figure(figsize=(25,8))
sns.lineplot(x = "tenure", y ="TotalCharges", data = data, color = "cyan")
plt.title("Comprison of Tenure with TotalCharges",fontsize=28)
```

# 4. Find out which contract is preferred by the senior citizen

```python
plt.title("Contract preference of senior citizen",fontsize=16)
sns.countplot(x='SeniorCitizen', hue = 'Contract', data = data)
```
